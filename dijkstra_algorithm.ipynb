```python
class Dijkstra:
    
    def __init__ (self, graph , start_node ,end_node):
        self.graph = graph
        self.start_node = start_node
        self.end_node = end_node
        # node stores the nodes of the graph
        self.nodes = list(graph.keys())
        
    def initialize(self):
        # at first all distance values are infinity
        self.distance = [float('inf')] * len(self.nodes)
        # set the distance value of the start point to 0
        index = self.nodes.index(self.start_node)
        self.distance[index]= 0
        # capture current node
        self.base_val = self.distance[index]
        # at fisrt all nodes are unvisited and False
        self.visited = [False] * len(self.nodes)
        # the start point is visited and True
        self.visited[index] = True
        # used for the final path
        self.route = {}
       
    def get_neighbours(self,node):
        return self.graph.get(node)
        
    def visited_node(self,node):
        index = self.nodes.index(node)
        self.visited[index] = True
        
    def check_visited(self,node):
        index = self.nodes.index(node)
        if self.visited[index] == True:
            return True
        else:
            return False
    
    def is_neighbour(self, node_1 ,node_2):
        result = False
        for key in self.get_neighbours(node_1).keys():
            if key == node_2:
                result = True
        return result
        
    def update_distance(self,current_node):  
        index = self.nodes.index(current_node)
        self.base_val = self.distance[index]
        
        # choose all neighbour of current node
        gn = self.get_neighbours(current_node)
        for key in gn.keys():
            print ('key =' )
            print (key)
            
            # check if neigh are unvisited then continue
            if self.check_visited(key) == False:
                index = self.nodes.index(key)
                dicvalue = gn[key]
                nodevalue = self.distance[index]
                # update distance based on min current dis and calculated dis
                
                if nodevalue > dicvalue + self.base_val:
                    self.distance[index] = (dicvalue + self.base_val)
                    self.route[key] = current_node
                print('route =' )
                print(self.route)
                print( 'distance =')
                print(self.distance)
                
        # we check all current node neigh --> make node visited
        self.visited_node(current_node)  
        
    
    def  get_min_val(self) :  
        mini = float('inf')
        
        # select unvisited with min distance node 
        for i in self.visited:
            if i == False:
                if self.distance[self.visited.index(i) ] < mini:
                    mini = self.distance[self.visited.index(i) ]
        print ('min =' )  
        print(mini)
        return (self.nodes[self.distance.index(mini)] )
        
        
    def next_node(self,current_node):
        self.update_distance(current_node)
        current_node = self.get_min_val()
        print('current_node = ')
        print(current_node)
        # check we reach end point or continue
        if current_node == self.end_node:
            return
        else:
            return self.next_node(current_node)
        
            
                    
    
    def  find_path(self): 
        self.initialize()
        self.next_node(self.start_node)
        lis= []
        
        lis.append(self.end_node)
        print (lis)
        
        for i in range (0,len(self.route)):
            if lis[i] != self.start_node:
                lis.append(self.route[lis[i]])
                print (lis)
            else :
                lis.reverse()
                return lis
        
      
```
